```python
# ðŸ“¦ Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# ðŸ“‚ Step 2: Load Dataset
df = pd.read_csv('accidents.csv')
print("âœ… Dataset Loaded Successfully")
print(df.head())

# ðŸ” Step 3: Basic Info
print("\nðŸ§¾ Dataset Info:")
print(df.info())
print("\nðŸ§¼ Null Values:\n", df.isnull().sum())

# ðŸ§¹ Step 4: Clean & Preprocess
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour
df['DayOfWeek'] = df['Start_Time'].dt.day_name()

# âœ… Remove missing weather or road condition if needed
df = df.dropna(subset=['Weather_Condition', 'Road_Condition'])

# ðŸ“Š Step 5: Visualizations

## 1. Accidents by Hour
plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data=df, palette='coolwarm')
plt.title('ðŸ•’ Number of Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.savefig('accidents_by_hour.png')
plt.show()

## 2. Accidents by Weather Condition
plt.figure(figsize=(12,5))
top_weather = df['Weather_Condition'].value_counts().nlargest(6).index
sns.countplot(data=df[df['Weather_Condition'].isin(top_weather)],
              x='Weather_Condition', palette='Set2')
plt.title('ðŸŒ¦ï¸ Accidents by Weather Condition')
plt.xlabel('Weather')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.savefig('accidents_by_weather.png')
plt.show()

## 3. Accidents by Road Condition
plt.figure(figsize=(8,4))
sns.countplot(x='Road_Condition', data=df, palette='muted')
plt.title('ðŸ›£ï¸ Accidents by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.savefig('accidents_by_road.png')
plt.show()

## 4. Accidents by Day of Week
plt.figure(figsize=(10,5))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='DayOfWeek', data=df, order=order, palette='viridis')
plt.title('ðŸ“† Accidents by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.savefig('accidents_by_day.png')
plt.show()

# ðŸ§  Step 6: Severity Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Severity', data=df, palette='magma')
plt.title('ðŸš¨ Severity of Accidents')
plt.xlabel('Severity Level')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('accidents_by_severity.png')
plt.show()

# ðŸ’¾ Step 7: Save Cleaned Dataset
df.to_csv("accidents_cleaned.csv", index=False)
print("âœ… Cleaned data saved as 'accidents_cleaned.csv'")

from google.colab import files
uploaded = files.upload()  # Upload accidents.csv
```
